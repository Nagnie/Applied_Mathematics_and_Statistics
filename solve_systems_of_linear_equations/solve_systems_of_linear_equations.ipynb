### Bài 1

```python
def Gauss_elimination(A):
    m = len(A)      
    n = len(A[0])  
    flag = 0        

    for r in range(m):
        
        if flag >= n:
            return A
        
        i = r

        while A[i][flag] == 0:
            i += 1
            if i == m:
                i = r
                flag += 1
                if n == flag:
                    return A
                
        A[i], A[r] = A[r], A[i]

        lv = A[r][flag]

        for i in range(len(A[i])):
            A[r][i] /= float(lv)
        
        for i in range(m):
            if i!= r:
                lv = A[i][flag]
                for j in range(len(A[i])):
                    A[i][j] -= lv * A[r][j]

        flag += 1
    return A

A = [[1, 2, -1, -1], [2, 2, 1, 1], [3, 5, -2, -1]]
print(Gauss_elimination(A))

A = [[1, -2, -1, 1], [2, -3, 1, 6], [3, -5, 0, 7], [1, 0, 5, 9]]
print(Gauss_elimination(A))
```

# Mô tả các hàm
Hàm **def Gauss_elimination(A)**: Chuyển đổi ma trận A thành ma trận bậc thang. Thuật toán được cài đặt như sau:
   
        • Khai báo các biến m và n lần lượt là số hàng và số cột của ma trận, flag là biến đánh dấu cột hiện tại đang xét.
        
        • Sử dụng vòng lặp lặp qua từng hàng của ma trận. Nếu biến flag lớn hơn/bằng n thì ta đã duyệt hết các cột, trả về ma trận A.
        
        • Trong vòng lặp while, tìm hàng đầu tiên mà phần tử ở cột flag khác 0. Nếu không tìm thấy, tăng flag lên 1 hoặc trả về ma trận nếu flag vượt quá số cột.
        
        • Hoán đổi vị trí của hai dòng vị trí i và r để đưa số hạng khác 0 lên trên cùng.
        
        • Lưu giá trị của số hạng hàng r cột flag là lv, chia các số hạng cùng hàng cho lv để biến số hạng đầu tiên thành 1.
        
        • Tới lần lượt các hàng tiếp theo, trừ các số hạng để đưa số hạng ở cột flag về 0.
        
        • Tăng flag để tiếp tục duyệt cột tiếp theo.

Hàm **def back_substitution (A)**: Tìm nghiệm của hệ phương trình tuyến tính với ma trận đầu vào là ma trận bậc thang đã rút gọn.

        • Khởi tạo một list để lưu trữ nghiệm
        
        • So sánh hạng của ma trận (A) và ma trận (A|B). Nếu hạng của ma trận (A) lớn hơn thì hệ phương trình vô nghiệm. Nếu hạng của hai ma trận bằng nhau và bằng số cột của ma trận thì hệ có một nghiệm xác định. Tìm và trả về nghiệm đó.
        
        • Trường hợp còn lại, phương trình vô số nghiệm, ta tìm và trả về dạng tổng quát của nghiệm.

Ở bài tập 3, dùng thư viện **numpy** để hỗ trợ tìm hạng của ma trận. Ta thấy dù dùng thư viện hay không thì kết quả vẫn như nhau.
