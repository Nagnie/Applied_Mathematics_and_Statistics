## Đề bài

File **"wine.csv"** là cơ sở dữ liệu đánh giá chất lượng của 1200 chai rượu vang theo thang điểm 1 - 10 dựa trên 11 tính chất khác nhau.

Xây dựng mô hình đánh giá chất lượng rượu sử dụng phương pháp hồi quy tuyến tính.

&emsp;&emsp;a. Sử dụng toàn bộ 11 đặc trưng đề bài cung cấp.

&emsp;&emsp;b. Sử dụng duy nhất 1 đặc trưng cho kết quả tốt nhất. (Gợi ý: Phương pháp Cross Validation)

&emsp;&emsp;c. Xây dựng một mô hình của riêng bạn cho kết quả tốt nhất.

## Giải thuật

**Bước 1:** Chuẩn bị dữ liệu

&emsp;&emsp;- Trích xuất ma trận $X$ từ dữ liệu, ngoại trừ biến mục tiêu **"quality"**.

&emsp;&emsp;- Trích xuất vectơ mục tiêu $y$ từ dữ liệu là cột **"quality"**.

&emsp;&emsp;- Thêm số hạng chặn (cột số $1$) vào đầu ma trận $X$.

**Bước 2:** Tính toán ma trận

&emsp;&emsp;- Tìm ma trận chuyển vị của $X$

&emsp;&emsp;- Tính tích ma trận $X^{T} X$, $X^{T} y$
    
**Bước 3:** Khởi tạo ma trận $X_{inv}$ là ma trận nghịch đảo của $X^{T} X$ bằng phép chuyển đổi Gauss

**Bước 4:** Tính hệ số hồi quy: 

&emsp;&emsp;Tính hệ số hồi quy $beta$ bằng cách nhân ma trận nghịch đảo $X_{inv}$ với $X^{T} y$.

**Bước 5:** Dự đoán và tính toán $MSE$ (Mean Squared Error):
$$
MSE = \frac{1}{n} \sum_{i=1}^n(y_{i} + \hat{y}_{i})^2
$$

**Bước 6:** Trả về kết quả $beta$ và $MSE$


### Câu a

```python
import pandas as pd
df = pd.read_csv('wine.csv')
df
```

```python
def Linear_Regression_a(s):
    match s:
        case 0:
            X = df.drop(columns=['\"quality\"'])
        case 1:
            X = df[['\"fixed acidity\"']]
        case 2:
            X = df[['\"volatile acidity\"']]
        case 3:
            X = df[['\"citric acid\"']]
        case 4:
            X = df[['\"residual sugar\"']]
        case 5:
            X = df[['\"chlorides\"']]
        case 6:
            X = df[['\"free sulfur dioxide\"']]
        case 7:
            X = df[['\"total sulfur dioxide\"']]
        case 8:
            X = df[['\"density\"']]
        case 9:
            X = df[['\"pH\"']]
        case 10:
            X = df[['\"sulphates\"']]
        case 11:
            X = df[['\"alcohol\"']]
    
    y = df['\"quality\"']
    X.insert(0, 'intercept', 1)


    X_transpose = X.T
    X_transpose_X = X_transpose.dot(X)
    X_transpose_y = X_transpose.dot(y)
    X_transpose_X_inv = pd.DataFrame(
        index=X_transpose_X.columns, 
        columns=X_transpose_X.index
    )

    X_transpose_X_inv[:] = 0

    n = len(X_transpose_X)

    A = X_transpose_X.copy()

    I = pd.DataFrame(
        [[1 if i == j else 0 for j in range(n)] for i in range(n)], 
        index=X_transpose_X.columns, 
        columns=X_transpose_X.index
    ).astype(float)
    
    for i in range(n):
        diag = float(A.iloc[i, i])
        A.iloc[i] /= diag
        I.iloc[i] /= diag
        for j in range(n):
            if i != j:
                ratio = A.iloc[j, i]
                A.iloc[j] -= ratio * A.iloc[i]
                I.iloc[j] -= ratio * I.iloc[i]

    X_transpose_X_inv = I
    beta = X_transpose_X_inv.dot(X_transpose_y)
    predictions = X.dot(beta)
    mse = ((y - predictions) ** 2).mean()
    return beta, mse

beta, mse = Linear_Regression_a(0)
print(beta)
print("MSE = ", mse)
print("quality = ", beta.iloc[0] + beta.iloc[1], "* fixed acidity + ", beta.iloc[2], "* volatile acidity + ", beta.iloc[3], "* citric acid + ", beta.iloc[4], "* residual sugar + ", 
      beta.iloc[5], "* chlorides + ", beta.iloc[6], "* free sulfur dioxide + ", beta.iloc[7], "* total sulfur dioxide + ", 
      beta.iloc[8], "* density + ", beta.iloc[9], "* pH + ", beta.iloc[10], "* sulphates + ", beta.iloc[11], "* alcohol")
```

### Câu b

```python
def Linear_Regression_b():
    m = []
    for i in range(1, 12):
        beta, mse = Linear_Regression_a(i)
        m.append(mse)

    min_mse = min(m)
    index = m.index(min_mse) + 1
    beta, mse = Linear_Regression_a(index)
    match index:
        case 1:
            X = "fixed acidity"
        case 2:
            X = "volatile acidity"
        case 3:
            X = "citric acid"
        case 4:
            X = "residual sugar"
        case 5:
            X = "chlorides"
        case 6:
            X = "free sulfur dioxide"
        case 7:
            X = "total sulfur dioxide"
        case 8:
            X = "density"
        case 9:
            X = "pH"
        case 10:
            X = "sulphates"
        case 11:
            X = "alcohol"
    
    return beta, X, m

beta, s, m = Linear_Regression_b()
print(beta)
print(s)
print("Đặc trưng cho kết quả tốt nhất: quality = ", beta.iloc[0] + beta.iloc[1], "*", s)
```

### Câu c

```python
def Linear_Regression_c():
    beta, mse = Linear_Regression_a(0)
    best_feature_index = beta[1:].abs().idxmax()
    feature_name = best_feature_index.replace('\"', '').strip()
    feature_index = list(df.columns).index(best_feature_index)
    beta_single, mse_single = Linear_Regression_a(feature_index)
    return beta_single, mse_single, feature_name

beta_single, mse_single, feature_name = Linear_Regression_c()
print(beta_single)
print("MSE = ", mse_single)
print("quality = ", beta_single.iloc[0], "+", beta_single.iloc[1],  "*", feature_name)
```

## Mô tả ý tưởng

**Câu a.**

Tạo hàm **def Linear_Regression_a(s):** với tham số đầu vào là $s$. Nếu $s$ bằng $0$ thì tìm mô hình sử dụng toàn bộ $11$ đặc trưng. Ngược lại sẽ tìm mô hình theo $1$ đặc trưng tương ứng (để ứng dụng cho câu b). Xây dựng hàm theo giải thuật ở trên.

**Câu b.** 

Đặc trưng cho kết quả tốt nhất sẽ là đặc trưng có $MSE$ nhỏ nhất. Như vậy, ta chạy một vòng lặp, mỗi lần lặp sẽ chạy hàm **def Linear_Regression_a(s):** để tìm $MSE$ tương ứng khi biểu diễn "quality" theo đặc trưng đó. Sau đó chọn ra $MSE$ có giá trị nhỏ nhất trong mảng và in ra màn hình.

**Câu c.** 

Dựa trên các hệ số hồi quy, ta có thể thấy mức độ ảnh hưởng của chúng lên chất lượng rượu. Như vậy, ta tìm ra đặc trưng có hệ số hồi quy lớn nhất và biểu diễn "quality" theo đặc trưng đó.
